```python
# --- Mount Google Drive ---
from google.colab import drive
drive.mount('/content/drive', force_remount=True)

# --- Import ---
import tensorflow as tf
import numpy as np
import pathlib
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
from tensorflow.keras.applications import EfficientNetB0, ResNet50
from tensorflow.keras.applications.efficientnet import preprocess_input as efficientnet_preprocess
from tensorflow.keras.applications.resnet50 import preprocess_input as resnet_preprocess

# --- Dataset Setup (Sama seperti sebelumnya) ---
data_dir = pathlib.Path("/content/drive/MyDrive/Dataset Gabungan")
IMG_SIZE = (224, 224)
BATCH_SIZE = 32
SEED = 123

raw_train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.3,
    subset="training",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)
class_names = raw_train_ds.class_names
num_classes = len(class_names)

raw_val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.3,
    subset="validation",
    seed=SEED,
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

val_batches = tf.data.experimental.cardinality(raw_val_ds)
val_size = val_batches // 2
val_ds = raw_val_ds.take(val_size)
test_ds = raw_val_ds.skip(val_size)

AUTOTUNE = tf.data.AUTOTUNE
train_ds = raw_train_ds.cache().shuffle(1000).prefetch(AUTOTUNE)
val_ds = val_ds.cache().prefetch(AUTOTUNE)
test_ds = test_ds.cache().prefetch(AUTOTUNE)

# --- Base Model 1: EfficientNetB0 ---
def build_efficientnet_base():
    base = EfficientNetB0(weights='imagenet', include_top=False, input_shape=IMG_SIZE+(3,))
    base.trainable = False
    return base

# --- Base Model 2: ResNet50 ---
def build_resnet_base():
    base = ResNet50(weights='imagenet', include_top=False, input_shape=IMG_SIZE+(3,))
    base.trainable = False
    return base

# Buat instance base models
effnet_base = build_efficientnet_base()
resnet_base = build_resnet_base()

# --- Stacking Ensemble Model ---

# Input untuk model ensemble
inputs = tf.keras.Input(shape=IMG_SIZE+(3,), name='input_layer')

# Jalur 1: EfficientNetB0
x1 = efficientnet_preprocess(inputs) # Preprocessing spesifik EfficientNet
x1 = effnet_base(x1, training=False)
x1 = tf.keras.layers.GlobalAveragePooling2D(name='effnet_gap')(x1)
x1 = tf.keras.layers.Dropout(0.4, name='effnet_dropout')(x1)
# *Tidak menggunakan layer Dense output di sini, karena ini akan digabungkan*

# Jalur 2: ResNet50
x2 = resnet_preprocess(inputs) # Preprocessing spesifik ResNet50
x2 = resnet_base(x2, training=False)
x2 = tf.keras.layers.GlobalAveragePooling2D(name='resnet_gap')(x2)
x2 = tf.keras.layers.Dropout(0.5, name='resnet_dropout')(x2)
# *Tidak menggunakan layer Dense output di sini, karena ini akan digabungkan*

# Menggabungkan output dari kedua base models (Stacking)
# Ini adalah tahap "meta-learner" atau "level 1 model"
concatenated = tf.keras.layers.Concatenate()([x1, x2])

# Meta-Learner (Simple Dense Layer)
# Layer ini mempelajari cara terbaik menggabungkan fitur dari kedua model dasar
x = tf.keras.layers.Dense(256, activation='relu', name='meta_dense_1')(concatenated)
x = tf.keras.layers.Dropout(0.3, name='meta_dropout')(x)
outputs = tf.keras.layers.Dense(num_classes, activation='softmax', name='output_layer')(x)

# Buat model Stacking
stacked_model = tf.keras.Model(inputs, outputs, name='Stacking_EfficientNetB0_ResNet50')

# Kompilasi
stacked_model.compile(optimizer='adam',
                      loss='sparse_categorical_crossentropy',
                      metrics=['accuracy'])

stacked_model.summary()

print("=== Training Stacking Ensemble Model ===")
# Latih model Stacking. Disarankan untuk menggunakan lebih sedikit epochs
# karena base models sudah pre-trained dan hanya melatih meta-learner.
history = stacked_model.fit(train_ds,
                            validation_data=val_ds,
                            epochs=10) # Mengurangi epoch menjadi 10 sebagai contoh

# --- Evaluation ---
print("\n=== Evaluation on Test Dataset ===")
stacked_model.evaluate(test_ds)

# --- Predict & Report ---
y_true = np.concatenate([y for x, y in test_ds], axis=0)
y_pred = np.argmax(stacked_model.predict(test_ds), axis=1)

print("\n=== Classification Report ===")
print(classification_report(y_true, y_pred, target_names=class_names))
print("\n=== Confusion Matrix ===")
print(confusion_matrix(y_true, y_pred))

# --- Plot history ---
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train')
plt.plot(history.history['val_accuracy'], label='Val')
plt.title('Accuracy'); plt.legend()
plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train')
plt.plot(history.history['val_loss'], label='Val')
plt.title('Loss'); plt.legend()
plt.show()


# ----------------------------- Grad-CAM Manual Input -----------------------------

import cv2

def tensor_to_uint8(img_tensor):
    img = img_tensor.numpy()
    if img.max() <= 1.0:
        img = (img * 255).astype('uint8')
    return img.astype('uint8')

def gradcam_for_ensemble(stack_model, effnet_base, resnet_base, img, class_index=None):

    img_tensor = tf.expand_dims(img, axis=0)

    # Preprocessing sesuai base model
    x_eff = efficientnet_preprocess(tf.cast(img_tensor, tf.float32))
    x_res = resnet_preprocess(tf.cast(img_tensor, tf.float32))

    # Ambil layer persis seperti arsitektur training
    effnet_gap = stack_model.get_layer('effnet_gap')
    effnet_dropout = stack_model.get_layer('effnet_dropout')
    resnet_gap = stack_model.get_layer('resnet_gap')
    resnet_dropout = stack_model.get_layer('resnet_dropout')
    meta_dense = stack_model.get_layer('meta_dense_1')
    meta_dropout = stack_model.get_layer('meta_dropout')
    output_layer = stack_model.get_layer('output_layer')

    with tf.GradientTape(persistent=True) as tape:
        # forward pass base models
        act1 = effnet_base(x_eff, training=False)
        act2 = resnet_base(x_res, training=False)

        tape.watch(act1)
        tape.watch(act2)

        # 🔥 forward pass harus sama persis seperti training
        gap1 = effnet_gap(act1)
        gap1 = effnet_dropout(gap1, training=False)

        gap2 = resnet_gap(act2)
        gap2 = resnet_dropout(gap2, training=False)

        concat = tf.concat([gap1, gap2], axis=-1)

        x = meta_dense(concat)
        x = meta_dropout(x, training=False)
        preds = output_layer(x)

        if class_index is None:
            class_index = tf.argmax(preds[0])

        class_score = preds[:, class_index]

    # Hitung gradien untuk kedua base model
    grads1 = tape.gradient(class_score, act1)
    grads2 = tape.gradient(class_score, act2)

    weights1 = tf.reduce_mean(grads1, axis=(1,2))
    weights2 = tf.reduce_mean(grads2, axis=(1,2))

    cam1 = tf.reduce_sum(tf.multiply(act1, weights1[:, None, None, :]), axis=-1)[0].numpy()
    cam2 = tf.reduce_sum(tf.multiply(act2, weights2[:, None, None, :]), axis=-1)[0].numpy()

    # Normalisasi
    heatmap = (cam1 + cam2) / 2
    heatmap = np.maximum(heatmap, 0)
    heatmap = heatmap / (heatmap.max() + 1e-8)
    heatmap = cv2.resize(heatmap, (IMG_SIZE[1], IMG_SIZE[0]))

    return heatmap, preds.numpy()[0], int(class_index.numpy() if hasattr(class_index, 'numpy') else class_index)

def overlay_heatmap_on_image(img, heatmap):
    heatmap = np.uint8(255 * heatmap)
    heatmap_color = cv2.applyColorMap(heatmap, cv2.COLORMAP_JET)
    heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
    overlay = cv2.addWeighted(img, 1, heatmap_color, 0.5, 0)
    return heatmap_color, overlay

# ----------------------------- USER INPUT IMAGE -----------------------------
img_path = "/content/drive/MyDrive/Dataset Gabungan/Ringworm/Copy of Salinan 438051747_429300536462648_7834621964132658771_n_jpg.rf.0db245ea8a8749a12a3f124c68dfe865.jpg"  # GANTI PATH GAMBAR

img = tf.keras.preprocessing.image.load_img(img_path, target_size=IMG_SIZE)
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_uint8 = img_array.astype("uint8")

pred = stacked_model.predict(tf.expand_dims(img_array, axis=0))
pred_class = np.argmax(pred)
pred_conf = pred[0][pred_class]

print(f"\nPrediksi: {class_names[pred_class]} ({pred_conf:.2f})")

heatmap, _, _ = gradcam_for_ensemble(stacked_model, effnet_base, resnet_base, img_array)

heatmap_rgb, overlay = overlay_heatmap_on_image(img_uint8, heatmap)

plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(img_uint8)
plt.title("Original"); plt.axis('off')

plt.subplot(1,3,2)
plt.imshow(heatmap, cmap='jet')
plt.title("Heatmap"); plt.axis('off')

plt.subplot(1,3,3)
plt.imshow(overlay)
plt.title(f"Overlay\n{class_names[pred_class]}"); plt.axis('off')

plt.show()
```
